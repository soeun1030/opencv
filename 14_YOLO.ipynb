```python
from ultralytics import YOLO
import cv2
import numpy as np

suzy_path = "images/suzy.jpg"
maru_path = "images/maru.jpg"
soeun_path = "images/soeun.png"
cars_path = "videos/cars.mp4"
```

```python
### YOLO 모델 불러오기
model = YOLO('yolo11n.pt')
img = cv2.imread(maru_path)

### 객체 인식, 신뢰도 0.5
results = model.predict(img, conf=0.5)

### 탐지 결과를 이미지 위에 그려줌
annnotated_frame = results[0].plot()

cv2.imshow("img", annnotated_frame)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(cars_path)
model = YOLO("yolo11n.pt")

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break
    
    results = model.predict(frame, conf=0.4)
    annnotated_frame = results[0].plot()

    cv2.imshow("video", annnotated_frame)

    if cv2.waitKey(1) == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
## 실습3. 캠화면에 적용

cap = cv2.VideoCapture(0)
model = YOLO("yolo11n.pt")


while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break
    
    results = model.predict(frame, conf=0.4)
    annnotated_frame = results[0].plot()

    cv2.imshow("video", annnotated_frame)

    if cv2.waitKey(1) == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
